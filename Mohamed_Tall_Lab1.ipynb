```python
#8 Create 30 qualitative datapoints
import pandas as pd 
import matplotlib.pyplot as plt
d2 = pd.Series([
    "Apple", "Banana", "Apple", "Orange", "Banana", "Grape", "Apple", "Apple", 
    "Orange", "Banana", "Grape", "Apple", "Orange", "Apple", "Banana", "Apple", 
    "Grape", "Apple", "Orange", "Banana", "Grape", "Banana", "Apple", "Grape", 
    "Apple", "Orange", "Banana", "Apple", "Grape", "Orange"
])
d2_counts = d2.value_counts()
plt.figure(figsize=(10, 5))
plt.bar(d2_counts.index, d2_counts.values, color=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title("Barplot of Favorite Fruits")
plt.xlabel("Fruit Type")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(d2_counts, labels=d2_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title("Pie Chart of Favorite Fruits")
plt.show()
```
